# Total data for Each Application


```python
df["social_mediadata"] = df["social_media_dl_(bytes)"] + df['social_media_ul_(bytes)']
df['youtube_data'] = df["youtube_dl_(bytes)"] + df["youtube_ul_(bytes)"]
df['netflix_data'] = df["netflix_dl_(bytes)"] + df["netflix_ul_(bytes)"]
df["gaming_data"] = df["gaming_dl_(bytes)"] + df["gaming_ul_(bytes)"]
df["google_data"] = df["google_dl_(bytes)"] + df["google_ul_(bytes)"]
df['email_data'] = df["email_dl_(bytes)"] + df["email_ul_(bytes)"]
df['other_data'] = df["other_dl_(bytes)"]+df["other_ul_(bytes)"]
df['total_data'] = df['total_dl_(bytes)'] + df['total_ul_(bytes)']
```

```python
user_data = df.groupby('msisdn/number')[['social_mediadata', 'google_data', 'email_data', 'youtube_data', 'gaming_data', 'other_data', 'total_data']].sum().nlargest(20, 'total_data')
```
